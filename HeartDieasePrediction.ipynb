```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split

smote = SMOTE()
x_test_resampled, y_test_resampled = smote.fit_resample(X_test, y_test)

x_tr2, x_tt2, y_tr2, y_tt2 = train_test_split(x_test_resampled, y_test_resampled, train_size=0.75, random_state=45)

```

```python
x_test_resampled.shape
```

```python
y_test_resampled.shape
```

```python
model2 = DecisionTreeClassifier(splitter='best',criterion = 'gini',max_depth= 15, min_samples_split = 2)
```

```python
model2.fit(x_tr2,y_tr2)
```

```python
model2.score(x_tr2,y_tr2)
```

```python
y_pr2 = model2.predict(x_tt2)
```

```python
accuracy_score(y_tt2, y_pr2)
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(40,30))
plot_tree(model2, filled=True, feature_names=X.columns, class_names=["No","Yes"])
plt.show()
```

```python
train_acc = model2.score(x_tr2, y_tr2)
test_acc = model2.score(x_tt2, y_tt2)

print("Training Accuracy:", train_acc)
print("Testing Accuracy:", test_acc)
```

```python
from sklearn.model_selection import GridSearchCV

# Define hyperparameter grid
param_grid = {
    'max_depth': [3, 5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'criterion': ['gini', 'entropy']
}

# Perform Grid Search
grid_search = GridSearchCV(DecisionTreeClassifier(), param_grid, cv=5, scoring='accuracy')
grid_search.fit(x_tr2, y_tr2)

# Print best parameters
print("Best Parameters:", grid_search.best_params_)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

y_pred2 = model2.predict(x_tt2)

# Confusion Matrix

cm3 = confusion_matrix(y_tt2, y_pred2)
print(confusion_matrix(y_tt2, y_pred2))

# Classification Report
print(classification_report(y_tt2, y_pred2))
```
